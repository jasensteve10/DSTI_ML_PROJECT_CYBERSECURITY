```python
target_column = 'Attack Type'

# Add binary missing indicators for the variables
df['Proxy_Info_Missing'] = df['Proxy Information'].isnull().astype(int)
df['Firewall_Logs_Missing'] = df['Firewall Logs'].isnull().astype(int)
df['IDS_IPS_Alerts_Missing'] = df['IDS/IPS Alerts'].isnull().astype(int)

# Group data by attack type and calculate the percentage of missing values for each variable
missing_analysis = df.groupby(target_column).agg({
    'Proxy_Info_Missing': 'mean',
    'Firewall_Logs_Missing': 'mean',
    'IDS_IPS_Alerts_Missing': 'mean'
}).reset_index()

# Rename columns for clarity
missing_analysis.rename(columns={
    'Proxy_Info_Missing': 'Proxy_Info_Missing_Percentage',
    'Firewall_Logs_Missing': 'Firewall_Logs_Missing_Percentage',
    'IDS_IPS_Alerts_Missing': 'IDS_IPS_Alerts_Missing_Percentage'
}, inplace=True)

# Display missing value percentages for each attack type
print("Missing Value Analysis by Attack Type:")
print(missing_analysis)

# Plotting the relationship between attack type and missing data
plt.figure(figsize=(12, 6))
missing_analysis.set_index(target_column).plot(kind='bar', stacked=False)
plt.title('Missing Value Percentages by Attack Type')
plt.ylabel('Missing Value Percentage')
plt.xlabel('Attack Type')
plt.legend(loc='upper right')
plt.grid(axis='y')
plt.show()
```

```python
def find_min_max(df, columns):
    """
    Function to find the min and max values for the given columns in a data frame.
    Returns:
    A dictionary with column names as keys and a tuple (min, max) as values.
    """
    result = {}
    
    for col in columns:
        if col in df.columns:
            min_value = df[col].min()
            max_value = df[col].max()
            result[col] = (min_value, max_value)
        else:
            print(f"Warning: Column '{col}' not found in the DataFrame.")
    
    return result
    
# Call the function
columns_to_check = ['Packet Length', 'Anomaly Scores']
min_max_values = find_min_max(df, columns_to_check)

print(min_max_values)
```

```python
def calculate_correlation(df):
    # Select only numeric columns from the DataFrame
    numeric_df = df.select_dtypes(include=['number'])
    
    # Calculate the correlation matrix
    correlation_matrix = numeric_df.corr()
    
    return correlation_matrix

correlation_matrix = calculate_correlation(df)
correlation_matrix
```

```python
def plot_frequencies(df, columns):
    # Set the style of seaborn
    sns.set(style="whitegrid")
    
    for column in columns:
        plt.figure(figsize=(10, 6))  # Set the figure size
        # Count the frequency of each category in the column
        frequency = df[column].value_counts()
        
        # Create a bar plot
        sns.barplot(x=frequency.index, y=frequency.values, palette='viridis')
        
        # Set the title and labels
        plt.title(f'Frequency of {column}', fontsize=16)
        plt.xlabel(column, fontsize=14)
        plt.ylabel('Frequency', fontsize=14)
        plt.xticks(rotation=45)  # Rotate x labels for better readability
        plt.show()  # Display the plot

plot_frequencies(df, ['Protocol', 'Attack Type', 'Severity Level'])
```

```python
# Group by date and Attack Type, then count occurrences
grouped_data = df.groupby(['time', 'Attack Type']).size().unstack(fill_value=0)

# Plotting the data
plt.figure(figsize=(12, 6))
grouped_data.plot(kind='line', marker='o')
plt.title('Attack Type Patterns Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Attacks')
plt.legend(title='Attack Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by date and Attack Type, then count occurrences
grouped_data = df.groupby(['time', 'Attack Type']).size().unstack(fill_value=0)

# Create subplots for each Attack Type
attack_types = grouped_data.columns
num_attack_types = len(attack_types)

# Set up the subplots
fig, axes = plt.subplots(num_attack_types, 1, figsize=(12, 6), sharex=True)

# Plot each Attack Type in a separate subplot
for ax, attack_type in zip(axes, attack_types):
    ax.plot(grouped_data.index, grouped_data[attack_type], marker='o', label=attack_type)
    ax.set_title(f'Number of {attack_type} Attacks Over Time')
    ax.set_ylabel('Number of Attacks')
    ax.legend()
    ax.grid()

# Set common x-axis label
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
